```python
import matplotlib.pyplot as plt
```

```python
x_cgpa = [2,3,4,5]
y_salary = [20000,40000,60000,80000]
plt.plot(x_cgpa,y_salary)
plt.xlabel("cgpa")
plt.ylabel("salary")
plt.title("cgpa vs salary")
plt.show()
```
